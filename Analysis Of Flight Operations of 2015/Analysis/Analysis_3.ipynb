# Analysis-3: 

Which day of the week experiences most arrival delay and which hour of the day experience most arrival delay?

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import timedelta
```

```python
# Reading data from 'Data' folder

cwd_path = os.getcwd() 
root_dir_path=os.path.abspath(os.path.join(cwd_path, os.pardir)) #To get the parent directory of cwd
root=root_dir_path.replace("\\", "/")
if not os.path.exists(root+"/"+'Data'): #To check for Data folder
    print('Path does not exist.')
else:
    if not os.path.exists(root+"/"+'Data'+"/"+'Flights'): #To check for Flights folder
        print('Path does not exist.')
    else:
        Path=root+"/"+'Data'+"/"+'Flights'
        if os.path.exists(Path+"/"+'flights.csv'): #To check for flights.csv
            Flights = pd.read_csv(Path+"/"+'flights.csv',sep=",")
        if os.path.exists(Path+"/"+'airlines.csv'): #To check for airlines.csv
            Airlines = pd.read_csv(Path+"/"+'airlines.csv',sep=",")
        if os.path.exists(Path+"/"+'airports.csv'): #To check for airports.csv
            Airports = pd.read_csv(Path+"/"+'airports.csv',sep=",")
```

```python
#Week name to all flights
Weeks= {1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday',7:'Sunday'}
Flights['WEEK_NAME'] = Flights['DAY_OF_WEEK'].apply(lambda x: Weeks[x])
Flights.head()
```

```python
#Fetching Scheduled Arrival Hour of the day
Flights['SCHEDULED_ARRIVAL_HOUR'] = Flights['SCHEDULED_ARRIVAL'].apply(lambda x: str(timedelta(minutes=x))[:-6][-2:].strip())
Flights.head()

#str(timedelta(minutes=100))[:-3] #Convert minutes to hour:minutes
#str(timedelta(minutes=100))[:-6] #Convert minutes to hour
```

```python
Flights_with_Arrival_Delay=Flights.dropna(subset=['ARRIVAL_DELAY']) #Dropping rows with null values

#Dataframe with average arrival delay by each day of week
df_arrival_delay_day_of_week =pd.DataFrame(Flights_with_Arrival_Delay.groupby(['DAY_OF_WEEK','WEEK_NAME']).mean()['ARRIVAL_DELAY'])
df_arrival_delay_day_of_week=df_arrival_delay_day_of_week.reset_index()
df_arrival_delay_day_of_week = df_arrival_delay_day_of_week.rename(columns={'ARRIVAL_DELAY': 'AVERGAE_ARRIVAL_DELAY(min)'})

df_arrival_delay_day_of_week.head()
```

```python
#Finding the most and least delay hours of day experiencing most and least delay

df_arrival_delay_day_of_week_Sorted=df_arrival_delay_day_of_week.sort_values(['AVERGAE_ARRIVAL_DELAY(min)'],ascending=[True])
df_arrival_delay_day_of_week_Sorted=df_arrival_delay_day_of_week_Sorted.reset_index(drop=True)
df_arrival_delay_day_of_week_Sorted


#Finding the most and least dealy hour of Least Delay Weekday
Least_Delay_WeekDay=df_arrival_delay_day_of_week_Sorted['WEEK_NAME'][0]
Least_Delay_WeekDay

df =pd.DataFrame(Flights_with_Arrival_Delay[Flights_with_Arrival_Delay['WEEK_NAME']==Least_Delay_WeekDay].groupby(['WEEK_NAME','SCHEDULED_ARRIVAL_HOUR']).mean()['ARRIVAL_DELAY'])
df=df.reset_index()
df=df.rename(columns={'SCHEDULED_ARRIVAL_HOUR':'HOUR_OF_THE_DAY','ARRIVAL_DELAY': 'AVERGAE_ARRIVAL_DELAY(min)'})

#Sorting
df=df.sort_values(['AVERGAE_ARRIVAL_DELAY(min)'],ascending=[True]).reset_index(drop=True)


#Finding the most and least dealy hour of Most Delay Weekday
Most_Delay_WeekDay=df_arrival_delay_day_of_week_Sorted['WEEK_NAME'][6]
Most_Delay_WeekDay

df1 =pd.DataFrame(Flights_with_Arrival_Delay[Flights_with_Arrival_Delay['WEEK_NAME']==Most_Delay_WeekDay].groupby(['WEEK_NAME','SCHEDULED_ARRIVAL_HOUR']).mean()['ARRIVAL_DELAY'])
df1=df1.reset_index()
df1=df1.rename(columns={'SCHEDULED_ARRIVAL_HOUR':'HOUR_OF_THE_DAY','ARRIVAL_DELAY': 'AVERGAE_ARRIVAL_DELAY(min)'})

#Sorting
df1=df1.sort_values(['AVERGAE_ARRIVAL_DELAY(min)'],ascending=[True]).reset_index(drop=True)

#Final Dataframe containing least and most delay hours of least and most dealyed Weekday
frames=[df.iloc[[0, -1]],df1.iloc[[0, -1]]]
df_final=pd.concat(frames)
df_final
```

```python
#Dataframe with average arrival delay by each hour of the day
df_arrival_delay_hour_of_day =pd.DataFrame(Flights_with_Arrival_Delay.groupby('SCHEDULED_ARRIVAL_HOUR').mean()['ARRIVAL_DELAY'])
df_arrival_delay_hour_of_day=df_arrival_delay_hour_of_day.reset_index()
df_arrival_delay_hour_of_day=df_arrival_delay_hour_of_day.rename(columns={'SCHEDULED_ARRIVAL_HOUR':'HOUR_OF_THE_DAY','ARRIVAL_DELAY': 'AVERGAE_ARRIVAL_DELAY(min)'})

#Sorting
#Converting datatype of HOUR_OF_THE_DAY to numeric for sorting
df_arrival_delay_hour_of_day[['HOUR_OF_THE_DAY']] = df_arrival_delay_hour_of_day[['HOUR_OF_THE_DAY']].apply(pd.to_numeric)
df_arrival_delay_hour_of_day=df_arrival_delay_hour_of_day.sort_values(['HOUR_OF_THE_DAY'],ascending=[True])
df_arrival_delay_hour_of_day.head()
```

```python
cwd = os.getcwd() #To access the path of current working directory
path=cwd+'/Analysis-3'
```

```python
#writing the output to csv

if not os.path.exists(path+'/'+'Analysis_3_Average_Arrival_Delay_by_Day_of_the_Week.csv'):
    df_arrival_delay_day_of_week.to_csv(path+'/'+'Analysis_3_Average_Arrival_Delay_by_Day_of_the_Week.csv', sep=',',index=False)
else:
    print('File already exists.')
    
if not os.path.exists(path+'/'+'Analysis_3_Least_and_Most_delay_hours_of_the_Least_and_Most_delay_experiencing_Day_of_the_Week.csv'):
    df_final.to_csv(path+'/'+'Analysis_3_Least_and_Most_delay_hours_of_the_Least_and_Most_delay_experiencing_Day_of_the_Week.csv', sep=',',index=False)
else:
    print('File already exists.')
    
if not os.path.exists(path+'/'+'Analysis_3_Average_Arrival_Delay_by_Hour_of_the_Day.csv'):
    df_arrival_delay_hour_of_day.to_csv(path+'/'+'Analysis_3_Average_Arrival_Delay_by_Hour_of_the_Day.csv', sep=',',index=False)
else:
    print('File already exists.')
```

```python
#Plot of Average Arrival Delay(minutes) for each Day of Week

plt.figure(figsize=(15,10))
ax=sns.barplot(x="WEEK_NAME", y="AVERGAE_ARRIVAL_DELAY(min)", data=df_arrival_delay_day_of_week,color='blue')
plt.title('Average Arrival Delay(minutes) for each Day of Week')
ax.set(xlabel='Week Name', ylabel='Average Arrival Delay(minutes)')
plt.xticks(rotation=20)
#plt.show()
plt.savefig(path+'/'+'Average_Arrival_Delay_by_Day_of_the_Week.jpeg', format="JPEG")
```

```python
# Plot of least most delay hours of the least and most delay experiencing weekday

#plt.figure(figsize=(25,25))
g = sns.factorplot(x='HOUR_OF_THE_DAY', y='AVERGAE_ARRIVAL_DELAY(min)',hue="WEEK_NAME",
                 data=df_final)
plt.title('Least and Most delayed Hours of the Least and Most delay experiencing Day of Week')
g.set(xlabel='Hour of the Day', ylabel='Average Arrival Delay(minutes)')
g.figsize=(15,10)
plt.xticks(rotation=20)
plt.legend(['Saturday', 'Monday'], loc='upper right')
#plt.show()
plt.savefig(path+'/'+'Least_and_Most_delay_hours_of_the_Least_and_Most_delay_experiencing_Day_of_the_Week.jpeg', format="JPEG")
```

```python
#Plot of Average Arrival Delay(minutes) for each Hour of Day

plt.figure(figsize=(15,10))
ax=sns.barplot(x="HOUR_OF_THE_DAY", y="AVERGAE_ARRIVAL_DELAY(min)", data=df_arrival_delay_hour_of_day,color='blue')
plt.title('Average Arrival Delay(minutes) for each Hour of Day')
ax.set(xlabel='Hour of the Day', ylabel='Average Arrival Delay(minutes)')
plt.xticks(rotation=20)
#plt.show()
plt.savefig(path+'/'+'Average_Arrival_Delay_by_Hour_of_the_Day.jpeg', format="JPEG")
```
