# Analysis-5:
What is the volume of filghts each Airline company operates over various months of an year?

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading data from 'Data' folder

cwd_path = os.getcwd() 
root_dir_path=os.path.abspath(os.path.join(cwd_path, os.pardir)) #To get the parent directory of cwd
root=root_dir_path.replace("\\", "/")
if not os.path.exists(root+"/"+'Data'): #To check for Data folder
    print('Path does not exist.')
else:
    if not os.path.exists(root+"/"+'Data'+"/"+'Flights'): #To check for Flights folder
        print('Path does not exist.')
    else:
        Path=root+"/"+'Data'+"/"+'Flights'
        if os.path.exists(Path+"/"+'flights.csv'): #To check for flights.csv
            Flights = pd.read_csv(Path+"/"+'flights.csv',sep=",")
        if os.path.exists(Path+"/"+'airlines.csv'): #To check for airlines.csv
            Airlines = pd.read_csv(Path+"/"+'airlines.csv',sep=",")
        if os.path.exists(Path+"/"+'airports.csv'): #To check for airports.csv
            Airports = pd.read_csv(Path+"/"+'airports.csv',sep=",")
```

```python
Flights.head()
```

```python
#Adding airline name to datafame
Airlines_dict = dict(zip(Airlines['IATA_CODE'],Airlines['AIRLINE']))
Flights['AIRLINE_NAME'] = Flights['AIRLINE'].apply(lambda x: Airlines_dict[x])
Flights.head()
```

```python
#Month name
Months = {1:'01-January',2:'02-February',3:'03-March',4:'04-April',5:'05-May',6:'06-June',7:'07-July',8:'08-August',9:'09-September',10:'10-October',11:'11-November',12:'12-December'}
Flights['MONTH_NAME'] = Flights['MONTH'].apply(lambda x: Months[x])
Flights.head()
```

```python
pvt_scheduled_airline_month = Flights.pivot_table(index="MONTH_NAME",columns="AIRLINE_NAME",values="SCHEDULED_DEPARTURE",aggfunc=lambda x: x.count())
pvt_scheduled_airline_month.head()
```

```python
cwd = os.getcwd() #To access the path of current working directory
path=cwd+'/Analysis-5'
```

```python
#Writing the output to csv

if not os.path.exists(path+'/'+'Analysis_5_Flight_Volume_of_each_Airlines_during_various_Months_of_Year.csv'):
    pvt_scheduled_airline_month.to_csv(path+'/'+'Analysis_5_Flight_Volume_of_each_Airlines_during_various_Months_of_Year.csv', sep=',',index=False)
else:
    print('File already exists.')
```

```python
#Plot
fig = plt.figure(figsize=(15,10))
ax=sns.heatmap(pvt_scheduled_airline_month, linecolor="w", linewidths=1);
plt.xticks(rotation=20)
plt.yticks(rotation=0)
plt.title('Flight Volume of each Airlines during various Months of Year(2015)')
plt.show()
plt.savefig(path+'/'+'Flight_Volume_of_each_Airlines_during_various_Months_of_Year(2015).jpeg', format="JPEG")
```
