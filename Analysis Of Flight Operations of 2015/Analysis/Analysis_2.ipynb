# Analysis-2:
* Flights scheduled, operated and cancellation trend by each month of 2015

* Most favourite and least favourite travelled state by each month of 2015


```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Reading data from 'Data' folder

cwd_path = os.getcwd() 
root_dir_path=os.path.abspath(os.path.join(cwd_path, os.pardir)) #To get the parent directory of cwd
root=root_dir_path.replace("\\", "/")
if not os.path.exists(root+"/"+'Data'): #To check for Data folder
    print('Path does not exist.')
else:
    if not os.path.exists(root+"/"+'Data'+"/"+'Flights'): #To check for Flights folder
        print('Path does not exist.')
    else:
        Path=root+"/"+'Data'+"/"+'Flights'
        if os.path.exists(Path+"/"+'flights.csv'): #To check for flights.csv
            Flights = pd.read_csv(Path+"/"+'flights.csv',sep=",")
        if os.path.exists(Path+"/"+'airlines.csv'): #To check for airlines.csv
            Airlines = pd.read_csv(Path+"/"+'airlines.csv',sep=",")
        if os.path.exists(Path+"/"+'airports.csv'): #To check for airports.csv
            Airports = pd.read_csv(Path+"/"+'airports.csv',sep=",")
```

```python
#Month name to all flights
Months = {1:'January',2:'February',3:'March',4:'April',5:'May',6:'June',7:'July',8:'August',9:'September',10:'October',11:'November',12:'December'}
Flights['MONTH_NAME'] = Flights['MONTH'].apply(lambda x: Months[x])
Flights.head()
```

```python
#Adding state name to Airports
States = States.rename(columns={'State': 'STATE'})
States_dict = dict(zip(States['Abbreviation'],States['STATE']))
Airports=Airports.dropna(subset=['STATE'])
Airports['STATE_NAME'] = Airports['STATE'].apply(lambda x: States_dict[x])
Airports.head()
```

```python
#Dropping rows where ORIGIN_AIRPORT and DESTINATION_AIRPORT is numeric

Flights['ORIGIN_AIRPORT'] = Flights['ORIGIN_AIRPORT'].str.replace('\d+', '')
Flights=Flights[Flights['ORIGIN_AIRPORT']!='']
Flights=Flights.dropna(subset=['ORIGIN_AIRPORT'])

Flights['DESTINATION_AIRPORT'] = Flights['DESTINATION_AIRPORT'].str.replace('\d+', '')
Flights=Flights[Flights['DESTINATION_AIRPORT']!='']
Flights=Flights.dropna(subset=['DESTINATION_AIRPORT'])
```

```python
#Adding origin state name to Flights
Airport_State_dict = dict(zip(Airports['IATA_CODE'],Airports['STATE_NAME']))
Flights['ORIGIN_AIRPORT_STATE'] = Flights['ORIGIN_AIRPORT'].apply(lambda x: Airport_State_dict[x])

#Adding destination state name to Flights
Airport_State_dict = dict(zip(Airports['IATA_CODE'],Airports['STATE_NAME']))
Flights['DESTINATION_AIRPORT_STATE'] = Flights['DESTINATION_AIRPORT'].apply(lambda x: Airport_State_dict[x])
Flights.head()
```

```python
#Flights scheduled, operated and cancellation trend by each month of 2015

Flights.CANCELLED.unique() #0 for not cancelled ,1 for cancelled
#Each flight is either cancelled or operated.

#Scheduled
Monthly_df = pd.DataFrame(Flights.groupby(['MONTH','MONTH_NAME']).count()['SCHEDULED_DEPARTURE'])

#Cancelled
Monthly_df['CANCELLED'] = pd.DataFrame(Flights.groupby(['MONTH','MONTH_NAME']).sum()['CANCELLED'])

#Operated=Scheduled-Cancelled
Monthly_df['OPERATED']=Monthly_df['SCHEDULED_DEPARTURE']-Monthly_df['CANCELLED']

#Operation%
Monthly_df['OPERATION_%']=Monthly_df['OPERATED']/Monthly_df['SCHEDULED_DEPARTURE']*100

#Cancellation%
Monthly_df['CANCELLATION_%']=Monthly_df['CANCELLED']/Monthly_df['SCHEDULED_DEPARTURE']*100

#Renaming the columns
Monthly_df = Monthly_df.rename(columns={'SCHEDULED_DEPARTURE': 'TOTAL_FLIGHTS_SCHEDULED','CANCELLED':'TOTAL_FLIGHTS_CANCELLED','OPERATED':'TOTAL_FLIGHTS_OPERATED'})
Monthly_df=Monthly_df.reset_index()
Monthly_df.head()
```

```python
cwd = os.getcwd() #To access the path of current working directory
path=cwd+'/Analysis-2'
```

```python
#Writing the output to csv

if not os.path.exists(path+'/'+'Analysis_2_Trend_of_Flights_Scheduled_Operated_Cancelled_by_Month.csv'):
    Monthly_df.to_csv(path+'/'+'Analysis_2_Trend_of_Flights_Scheduled_Operated_Cancelled_by_Month.csv', sep=',',index=False)
else:
    print('File already exists.')
```

```python
#Plot
ax1=sns.pointplot(x="MONTH_NAME", y="TOTAL_FLIGHTS_SCHEDULED", data=Monthly_df, figsize=(15,10),color='Blue',label='Scheduled')
ax1=sns.pointplot(x="MONTH_NAME", y="TOTAL_FLIGHTS_OPERATED", data=Monthly_df, figsize=(15,10),color='Green',label='Operated')
ax1=sns.pointplot(x="MONTH_NAME", y="TOTAL_FLIGHTS_CANCELLED", data=Monthly_df, figsize=(15,10),color='Red')

plt.title('Trend of Flight Scheduling, Operation  and Cancellation by each Month')
ax1.set(xlabel='Month Name', ylabel='Total Number of Flights')

ax1.legend(handles=ax1.lines[::len(Monthly_df)+1], labels=["Scheduled","Operated","Cancelled"])
plt.xticks(rotation=20)
plt.show()

plt.savefig(path+'/'+'Trend_of_Flights_Scheduled_Operated_Cancelled_by_Month.jpeg', format="JPEG")
```

```python
#Least-popular destination state

Destination_Ranking_Least_Favourite = pd.DataFrame(Flights.groupby(['MONTH_NAME','DESTINATION_AIRPORT_STATE','ORIGIN_AIRPORT_STATE'])['FLIGHT_NUMBER'].count().reset_index(name='TOTAL_FLIGHT_COUNT'))
Destination_Ranking_Least_Favourite.head()

#Flights by Month, Destination_Airport_State
df1=pd.DataFrame(Destination_Ranking_Least_Favourite.groupby(['MONTH_NAME','DESTINATION_AIRPORT_STATE'])['TOTAL_FLIGHT_COUNT'].min().reset_index(name='TOTAL_FLIGHT_COUNT'))
#df1.head()

#Inner join
df2=pd.merge(Destination_Ranking_Least_Favourite,df1,how='inner')
#df2.head()

#Flights by Month
df3=pd.DataFrame(df2.groupby(['MONTH_NAME'])['TOTAL_FLIGHT_COUNT'].max().reset_index(name='TOTAL_FLIGHT_COUNT'))
#df3.head()

#Inner join
#Least favourite destination by month
Destination_Ranking_Least_Favourite=pd.merge(df2,df3,how='inner')
Destination_Ranking_Least_Favourite

#Writing the output to csv
if not os.path.exists(path+'/'+'Analysis_2_Least_Favourite_Destination_State_by_Month.csv'):
    Destination_Ranking_Least_Favourite.to_csv(path+'/'+'Analysis_2_Least_Favourite_Destination_State_by_Month.csv', sep=',',index=False)
else:
    print('File already exists.')
```

```python
#Most favourite destination by Month

Destination_Ranking_Most_Favourite = pd.DataFrame(Flights.groupby(['MONTH_NAME','DESTINATION_AIRPORT_STATE','ORIGIN_AIRPORT_STATE'])['FLIGHT_NUMBER'].count().reset_index(name='TOTAL_FLIGHT_COUNT'))
#Destination_Ranking.head()

df1=pd.DataFrame(Destination_Ranking_Most_Favourite.groupby(['MONTH_NAME','DESTINATION_AIRPORT_STATE'])['TOTAL_FLIGHT_COUNT'].max().reset_index(name='TOTAL_FLIGHT_COUNT'))
df1.head()

df2=pd.merge(Destination_Ranking_Most_Favourite,df1,how='inner')
df2.head(20)

df3=pd.DataFrame(df2.groupby(['MONTH_NAME'])['TOTAL_FLIGHT_COUNT'].max().reset_index(name='TOTAL_FLIGHT_COUNT'))
df3.head()

Destination_Ranking_Most_Favourite=pd.merge(df2,df3,how='inner')
Destination_Ranking_Most_Favourite

#Writing the output to csv
if not os.path.exists(path+'/'+'Analysis_2_Most_Favourite_Destination_State_by_Month.csv'):
    Monthly_df.to_csv(path+'/'+'Analysis_2_Most_Favourite_Destination_State_by_Month.csv', sep=',',index=False)
else:
    print('File already exists.')
```
